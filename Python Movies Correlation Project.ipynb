```python
# Drop the na values
df = df.dropna()
```

```python
# check the data again. This time I'm interested in data types
df.info()
```

```python
# Changing budget and gross revenue to ints 
df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
```

```python
df.sort_values("gross", ascending = False)
```

```python
# Scatterplot with budget vs gross revenue

plt.scatter(x=df["budget"], y=df["gross"])
plt.title("Budget vs Gross Earnings")
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.show()
```

```python
# Plot the budget vs gross using seborn

sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Start looking at correlation
df.corr(numeric_only=True)
```

```python
# Using a heatmap to show the correlations for columns with numeric values
# We can see a high correlation (0.74) between budget and gross as predicted
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(df.corr(numeric_only=True), annot= True, cmap = "Reds")
plt.title("Correlation for Numeric Values")
plt.show()
```

```python
# Looking at Company which is non numeric
df.info()
```

```python
df_numerized = df

for col in df_numerized.columns:
    if(df_numerized[col].dtype == "object"):
        df_numerized[col] = df_numerized[col].astype("category")
        df_numerized[col] = df_numerized[col].cat.codes
        
df_numerized.head()
```

```python
# Checking the correlations now we have numerized the company values

plt.figure(figsize=(12,8))
sns.set(font_scale=0.8)
sns.heatmap(df_numerized.corr(numeric_only=True), annot= True, cmap = "Reds")
plt.title("Correlation for Numeric Values")
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Find only the pairs where correlation is greater than 0.5
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
